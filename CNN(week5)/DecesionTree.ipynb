Kiểm tra tập dữ liệu Titanic

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import DecesionTree as DT
```

```python
#Tải dữ liệu Titanic
url = "https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv"
data = pd.read_csv(url)
data.head()
```

```python
data.info()
```

```python
data.describe()
```

```python
plt.figure(figsize=(8,5))
sns.heatmap(data.isnull(), cbar=False, cmap="viridis")
plt.title("Missing Values in Titanic Dataset")
plt.show()
```

```python
plt.figure(figsize=(8,4))
sns.countplot(x="Survived", data=data)
plt.title("Survival Distribution (0 = Died, 1 = Survived)")
plt.show()

plt.figure(figsize=(8,4))
sns.countplot(x="Pclass", hue="Survived", data=data)
plt.title("Survival by Passenger Class")
plt.show()

plt.figure(figsize=(8,4))
sns.boxplot(x="Survived", y="Age", data=data)
plt.title("Age Distribution by Survival")
plt.show()
```

Suy ra các cột có ảnh hưởng mạnh đến yếu tố sống sót của nạn nhân là:
Pclass, Sex, Age, SibSp, Parch, Fare, Embarked.

```python
model = DT.TitanicDecesionTree(url, max_depth=4)
model.load_data()
model.train_model()
model.evaluate()
```

```python
# Minh họa cây quyết địnhcd
model.draw_decision_tree()
```
