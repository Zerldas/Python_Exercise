MATPLOTLIB

```python
# Biểu đồ phân tán

df.plot(kind='scatter', x='sepal_length', y='sepal_width', title='Sepal Length vs Sepal Width')
plt.show()
```

```python
# Biểu đồ phân bố của petal_length

df['petal_length'].hist(bins=20)
plt.title('Distribution of Petal Length')
plt.xlabel('Petal Length')
plt.ylabel('Frequency')
plt.show()
```

```python
# Biểu đồ cột

df.groupby('species')['sepal_length'].mean().plot(kind='bar', title='Average Sepal Length by Species')
plt.ylabel('Average Sepal Length')
plt.show()
```

```python
# Biều đò hợp 

df.boxplot(column='sepal_length', by='species')
plt.title('Boxplot of Sepal Length by Species')
plt.suptitle('')
plt.xlabel('Species')
plt.ylabel('Sepal Length')
plt.show()
```

```python
# Biều đồ đường của 30 mẫu đầu tiên

df.head(30)['sepal_length'].plot(kind='line', title='Sepal Length of First 30 Samples')
plt.xlabel('Sample Index')
plt.ylabel('Sepal Length')
plt.show()
```

```python
# Biều đồ cột chồng

df.groupby('species').mean().plot(kind='bar', stacked=True, title='Mean Values by Species (Stacked)')
plt.ylabel('Average Value')
plt.show()
```

```python
# Biều đồ tròn

df['species'].value_counts().plot(kind='pie', autopct='%1.1f%%', title='Proportion of Species')
plt.ylabel('')
plt.show()
```
